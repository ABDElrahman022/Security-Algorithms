# Caesar Cipher is simple encryption Technique 
## Work by Substitution ( Shifting the letters )

```python
import re
```

```python
print("""ATTENTION !!
Input the message using alphabetical letters only""")

while True:
    
    plainText = input("Enter your message ")
    if plainText and re.match("^[a-zA-Z\s]+$", plainText):
        break
    
    else:
        print("Invalid input. Please enter a message using alphabetical letters only.")
```

```python
while True:
    try:
        k = int(input("Enter the key "))
        break
    except ValueError:
        print("Invalid input. Please enter an integer value for the key")
    
```

# 

```python
def cipher(p,k):
    cipher_message=""
    for char in p:
        if char.isupper():
            i = (ord(char)+k-ord('A')) % 26 + ord('A')
            cipher_message += chr(i)
        elif char.islower():
            i = (ord(char)+k-ord('a')) % 26 + ord('a')
            cipher_message += chr(i)
        else :
            cipher_message += char
    return cipher_message
```

```python
c=cipher(plainText,k)
print(c)
```

```python
def original(c,k):
    original_message=""
    for char in c:
        if char.isupper():
            i = (ord(char)-k-ord('A')) % 26 + ord('A')
            original_message += chr(i)
        elif char.islower():
            i = (ord(char)-k-ord('a')) % 26 + ord('a')
            original_message += chr(i)
        else :
            original_message += char
    return original_message
```

```python
original(c,k)
```
